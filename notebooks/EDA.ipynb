## 3.1. Agregación de interacciones y pagos a nivel -Usuario-

> ***`dfinteractions_data`***

Nuevo dataframe: `dfinteractions_peruser`, con las columnas:

- UserID
- total_interactions
- number_helprqs
- number_contentviews
- mostly_viewedgenre
- first_contentviewdate
- last_contentviewdate
- first_helprqsdate
- last_helprqsdate

```python
# Creamos columnas temporales con la fecha solo si cumple la condición

dfinteractions_data['Date_ContentView'] = dfinteractions_data['InteractionDate'].where(dfinteractions_data['InteractionType'] == 'Content View', pd.NaT)
dfinteractions_data['Date_HelpRequest'] = dfinteractions_data['InteractionDate'].where(dfinteractions_data['InteractionType'] == 'Help Request', pd.NaT)
```

```python
# 1. Creamos una columna temporal que concatene el género con el número de vistas, solo para las filas que son 'Content View'
dfinteractions_data['Interaccion_Genero'] = 'number_views_genre' + dfinteractions_data['ContentGenre'].astype(str)

# Si volviste a correr la celda, eliminamos las columnas viejas para que no choquen
dfinteractions_data = dfinteractions_data.drop(columns=[c for c in dfinteractions_data.columns if 'number_views_genre' in c], errors='ignore')

# 2. Creamos columnas booleanas (0 y 1) SOLO para las filas que son 'Content View'
# Las filas que NO son 'Content View' se llenan con 0 de forma segura
dummies = pd.get_dummies(dfinteractions_data.loc[dfinteractions_data['InteractionType'] == 'Content View', 'Interaccion_Genero']).astype(int)
dfinteractions_data = dfinteractions_data.join(dummies)
dfinteractions_data[dummies.columns] = dfinteractions_data[dummies.columns].fillna(0).astype(int)

# 3. Guardamos los nombres de las nuevas columnas dinámicas para sumarlas en el agg del groupby
columnas_genero = list(dummies.columns)

# Creamos las agregaciones de género dinámicamente con un ciclo for (comprensión de diccionario)
dinamicas = {col: (col, 'sum') for col in columnas_genero}
```

```python
dfinteractions_peruser = dfinteractions_data.groupby("UserID").agg(
    total_interactions = ("InteractionDate", "count"),
    number_helprqs = ('InteractionType', lambda x: (x == 'Help Request').sum()),
    number_contentviews = ('InteractionType', lambda x: (x == 'Content View').sum()),
    first_contentviewdate=('Date_ContentView', 'min'),
    last_contentviewdate=('Date_ContentView', 'max'),
    first_helprqsdate=('Date_HelpRequest', 'min'),
    last_helprqsdate=('Date_HelpRequest', 'max'),
    **dinamicas # Agrega dinámicamente las columnas de género al agg del groupby
).reset_index()

# El operador ** (desempaquetado de diccionarios) toma las claves y valores del diccionario "dinamicas" 
# y se los entrega a .agg() como si los hubieras escrito manualmente uno por uno dentro de los paréntesis
```

## 3.2. merge de dataframes

> 1. Usar `functools.reduce` (La forma "profesional")

Esta es la forma estándar de unir una lista de DataFrames. reduce toma una función (en este caso `pd.merge`) y la aplica sucesivamente a los elementos de tu lista.

```python
from functools import reduce
```

```python
# Lista de los DataFrames
dfs = [dfusers_data, # datos de usuarios
       dfsubscriptions_data, # datos de suscripciones
       dfpayments_peruser, # datos de pagos por usuario
       dfcac_merged_updated_data, # datos de Costo de adquisición CAC
       dfinteractions_peruser] # datos de interacciones por usuario (vistas y help requests)

# Unirlos todos usando una columna común ('UserID', partiendo del data frame de usuarios)
df_netflixmerge = reduce(lambda left, right: pd.merge(left, right, on='UserID', how='left'), dfs)
```

# 4. Feature Engineering

> **Nuevas variables:**

- `lifetimevalue` = (total_payments)*(last_payment_amount) -> Dado que no cambiaron los valores pagados en el tiempo.
- `ltvcac_ratio` = round(lifetimevalue / CAC, 2)
- `netcustomervalue` = lifetimevalue - CAC
- `churned` = 0 if Cancellation == 'no cancellation' else 1

> **Nuevas variables dummies (*one hot encoding*)**

- `gender`
- `location`
- `CancelReason`
- `PlanType`

```python
# Creamos una copia del df final para hacerle transformaciones sin perder el original

df_netflixfinal = df_netflixmerge.copy()

df_netflixfinal.columns
```

```python
df_netflixfinal["CancelReason"].unique()
```

```python
# Nuevas variables

try:
    df_netflixfinal["lifetimevalue"] = round(df_netflixfinal["total_payments"] * df_netflixfinal["last_payment_amount"], 2)

    df_netflixfinal["ltvcac_ratio"] = round(df_netflixfinal["lifetimevalue"] / df_netflixfinal["CAC"], 2)

    df_netflixfinal["netcustomervalue"] = round(df_netflixfinal["lifetimevalue"] - df_netflixfinal["CAC"], 2)

    df_netflixfinal["churned"] = np.where(df_netflixfinal["CancelReason"] == "No cancellation", 0, 1)

    print("✅ Nuevas variables creadas exitosamente.")

except KeyError as e:
    print(f"🚩 Error: La columna {e} no se encuentra en el DataFrame. Asegúrate de que todas las columnas necesarias estén presentes.")
```

```python
df_netflixfinal.columns
```

> **OneHotEncoding**

```python
# One hot encoding

from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

# 1. Definir las columnas que queremos transformar
categorical_features = ['Gender', 'Location', 'CancelReason', 'PlanType']

# 2. Configurar el transformador

preprocessor = ColumnTransformer(
    transformers=[
        ('cat', OneHotEncoder(sparse_output=False, # para que devuelva un array denso y no una matriz dispersa
                              handle_unknown='ignore', # para que no de error si aparece una categoría nueva en el futuro
                              drop='first'), # para evitar la multicolinealidad, se puede elegir una categoría de referencia para cada variable y eliminarla del resultado final
                              categorical_features) # lista de columnas a transformar
    ],
    remainder='passthrough', # Mantiene las columnas que no fueron transformadas
    verbose_feature_names_out = False # Para que los nombres de las columnas salgan sin el prefijo del transformador ('cat__', 'remainder__')
)

# 3. Aplicar la transformación
# Esto devuelve un array de NumPy
data_encoded = preprocessor.fit_transform(df_netflixfinal)

# 4. Convertir de nuevo a DataFrame para mantener los nombres de las columnas
# Obtenemos los nuevos nombres de las columnas generadas
feature_names = preprocessor.get_feature_names_out()

df_final = pd.DataFrame(data_encoded, columns=feature_names)
```
